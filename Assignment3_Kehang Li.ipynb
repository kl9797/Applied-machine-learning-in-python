```python
def confusion_matrix_plot(cm):
    
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.imshow(cm)
    ax.grid(False)
    ax.set_xlabel('Predicted outputs', fontsize= 18, color='black')
    ax.set_ylabel('Actual outputs', fontsize= 18, color='black')
    ax.xaxis.set(ticks=range(3))
    ax.yaxis.set(ticks=range(3))
    ax.set_ylim(2.5, -0.5)
    for i in range(3):
        for j in range(3):
            ax.text(j, i, cm[i, j], ha='center', va='center', color='white')
    plt.show()
```

```python
lst = []
lst1 = []
lst2 = []
for i in [x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11]:
    X_train, X_test, y_train, y_test = train_test_split(i, y, test_size=0.2, random_state=0)
    
    model = LogisticRegression(solver='lbfgs', multi_class='ovr',random_state=0)
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    model.score(X_train, y_train)
    model.score(X_test, y_test)
    
    cm = confusion_matrix(y_test, y_pred)
    lst.append(model.score(i, y))
    lst1.append(model.score(X_test, y_test))
    lst2.append(np.max(model.n_iter_))
    
    print("-------- Multiple logistic regression model------------------- ")
    print("Accuracy of the model: {:.3f}".format(model.score(i, y)))
    print("iterations to convergence: {:.1f}".format(np.max(model.n_iter_)))
    print(classification_report(y, model.predict(i)))
    
    
    confusion_matrix_plot(cm)
    
```

```python
# Among all the 11 cases, case 6 has highest model accuracy, 
# select case 6 as the features set for the regularization parameter C grid

x6 = data.iloc[:,[2,3]].values
X_train, X_test, y_train, y_test = train_test_split(x6, y, test_size=0.2, random_state=0)

model = LogisticRegression(solver='lbfgs', multi_class='ovr',random_state=0)
modelF = model.fit(X_train, y_train)
y_predF = modelF.predict(X_test)

C = [0.1, 1, 10, 100, 1000]
hyperF = dict(C = C)

gridF = GridSearchCV(model, hyperF, cv = 3, verbose = 1, 
                      n_jobs = -1)

bestF = gridF.fit(X_train, y_train)
```

### 5.2 Regularization path of L1- Logistic Regression and L2- Logistic Regression

```python
from time import time
from sklearn import linear_model
from sklearn import datasets
from sklearn.svm import l1_min_c

x6 = data.iloc[:,[2,3]].values
X = x6

cs = l1_min_c(X, y, loss='log') * np.logspace(0, 7, 16)

print("Computing regularization path ...")
start = time()
clf = LogisticRegression(penalty='l1', solver='liblinear', multi_class='ovr',
                                      tol=1e-6, max_iter=int(1e6),
                                      warm_start=True,intercept_scaling=10000.)

clf2 = LogisticRegression(penalty='l2', solver='lbfgs', multi_class='ovr',
                                      tol=1e-6, max_iter=int(1e6),
                                      warm_start=True,
                                      intercept_scaling=10000.)
```

```python
def logistic_regression_path_plot(clf):
    coefs_ = []
    for c in cs:
        clf.set_params(C=c)
        clf.fit(X, y)
        coefs_.append(clf.coef_.ravel().copy())
    print("This took %0.3fs" % (time() - start))

    coefs_ = np.array(coefs_)
    plt.plot(np.log10(cs), coefs_, marker='o')
    ymin, ymax = plt.ylim()
    plt.xlabel('log(C)')
    plt.ylabel('Coefficients')
    plt.title('Logistic Regression Path')
    plt.axis('tight')
    plt.show()
```

```python
# Logistic regression path plot for L1 regularization
logistic_regression_path_plot(clf)
```

```python
# Logistic regression path plot for L2 regularization
logistic_regression_path_plot(clf2)
```

## 6. Discuss your findings. Does using more dimensions help when trying to classify the data in this dataset? How important is regularization in these cases?

#### From the perspectives of model accuracy and macro-average precision, adding more dimensions didn't help to improve the model performance. In 2-dimensional feature set, the prediction using training data of variables sepal length and sepal width has highest accuracy (AUC = 0.96), whereas the accuracy of 3-dimensional training data are both lower than that.
#### Regularization is important to avoid overfitting problem, it can reduce the model variance without significant increase in bias. The equations of L1 and L2 regularizations are showed as following:

 \begin{align*}
{\bf \hat{\beta}_{\text{L1}}}
= \arg\min_{\bf \beta} \big( \sum_{i=1}^{n} (y_i- (\beta_0 + \beta_1 x_{i, 1} + ... + \beta_p x_{i,p}))^2
  + \lambda \sum_{j=0}^{p} | \beta_j | \big)
\\
\tag{1}
\end{align*}

\begin{equation*}
{\bf \hat{\beta}_{\text{L2}}}
= \arg\min_{\bf \beta} \big( \sum_{i=1}^{n} (y_i-(\beta_0 + \beta_1 x_{i,1} + ... + \beta_p x_{i,p}))^2
  + \lambda \sum_{j=0}^{p} | \beta_j | ^2 \big)
\tag{2}
\end{equation*}

#### As the plots of L1 and L2 regularization paths showed, the paths for each dimension's coefficient are similar in two plots. When the regularization is strong, log(C) is negative, all the coefficients are approximate to 0, and vary to non-zero values when regularization become looser. 
#### Regularization can improve the accuracy in this logistic model, which is proved in grid search, the accuracy of the model with parameter C = 10 is higher than the model without regularization.
